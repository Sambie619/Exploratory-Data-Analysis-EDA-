```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("world_population_data.csv")
```

```python
#visualization 
```

```python
sns.histplot(df["2023 population"],bins=10,kde=True)
plt.title("2023 population frequency distribution")
plt.xlabel("population")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby("continent")["2023 population"].mean().sort_values(ascending=False)
```

```python
#2 least populated continent
df.groupby("continent")["2023 population"].mean().sort_values(ascending=False).tail(2)
```

```python
top10countries=df.sort_values(by="2023 population",ascending=False).head(10)
```

```python
top10countries
```

```python
sns.barplot(x="2023 population",y="country",data=top10countries)
plt.title("top 10 countries by polulation in 2023")
plt.xlabel("population")
plt.ylabel("country")
plt.show()
```

```python
df.columns
```

```python
df.rename(columns={"density (km²)":"density (km2)"},inplace=True)
```

```python
sns.scatterplot(x="2023 population",y="density (km2)",data=df)
plt.title("population density of 2023")
plt.xlabel("2023 population")
plt.ylabel("Density")
plt.show()
```

```python
df.rename(columns={"area (km²)":"area (km2)"},inplace=True)
```

```python
df.groupby("continent")["area (km2)"].sum().sort_values(ascending=False)
```

```python
sns.violinplot(x="continent",y="density (km2)",data=df,palette="viridis")
plt.title("Population density distribution across continents")
plt.xlabel("Continent")
plt.ylabel("Population Density")
plt.show()
```
