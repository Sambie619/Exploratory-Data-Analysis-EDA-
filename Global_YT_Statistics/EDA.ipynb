```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#loading the dataset 
df=pd.read_csv("Global YouTube Statistics.csv",encoding="latin1")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
pd.set_option("display.float_format","{:.0f}".format)
```

```python
#Subscribers count by category 
Sub_count_by_category=df.groupby("category")["subscribers"].mean()
Sub_count_by_category
```

```python
Sub_count_by_category.sort_values(ascending=False,inplace=True)
```

```python
Sub_count_by_category.plot(kind="bar")
plt.title("Sub Count by Category")
plt.xlabel("Caategory")
plt.xticks(rotation=90)
plt.ylabel("Sub Count")
plt.show()
```

```python
df.columns
```

```python
#Channel type counts 
channel_type_count=df["channel_type"].value_counts()
channel_type_count
```

```python
#Pie chart of top 10 channel types 
channel_type_count.head(10).plot(kind="pie",autopct="%1.1f%%")
plt.title("Distribution of Channel types")
plt.show()
```

```python
#Scatter Plot of subscribers and video views 
sns.scatterplot(data=df,x="subscribers",y="video views")
plt.title("Subscribers vs video views")
plt.show()
```

```python
#Subs count by country 
Sub_count_by_country=df.groupby("Country")["subscribers"].sum().sort_values(ascending=False)
Sub_count_by_country
```

```python
Sub_count_by_country.head(10).plot(kind="bar")
plt.title("Sub count by country")
plt.xlabel("Country")
plt.xticks(rotation=90)
plt.ylabel("Sub count")
plt.show()
```

```python
#boplot of highest monthly earnings per channel type
sns.boxplot(x="channel_type",y="highest_monthly_earnings",data=df)
plt.title("Highest monthly earnings by channel type")
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby(["category","Country"])["subscribers"].sum().unstack().plot(kind="bar",stacked=True)
plt.title("Subscribers by category and country")
plt.xlabel("Category")
plt.ylabel("Subscribers")
plt.legend(title="Country",bbox_to_anchor=(1.05,1),loc="upper left")
plt.xticks(rotation=90)
plt.show()
#this chart shows the subscribers count per category and country ,and country is stacked in each 
#category type 
```

```python
df.columns
```

```python
plt.scatter(df["Gross tertiary education enrollment (%)"],df["subscribers"],alpha=0.5)
plt.title("Scatterplot of gross tertiary education enrollment vs subscribers")
plt.xlabel("gross tertiary education enrollment")
plt.ylabel("subscribers")
plt.show()
```

```python
df.groupby("channel_type")["Unemployment rate"].mean().sort_values(ascending=False).plot(kind="bar")
plt.title("Unemployment rate by channel type")
plt.xlabel("Channel type")
plt.ylabel("Unemployment rate")
plt.show()
#bar chart showing unemployment rate by channel type
```

```python
df.groupby(["created_year","channel_type"])["subscribers"].mean().unstack().tail().plot(kind="bar",width=1)
plt.title("Avg subscribers by created year and channel types")
plt.xlabel("Created year and channel types")
plt.ylabel("Subscribers")
plt.legend(title="Channel Type",bbox_to_anchor=(1.05,1),loc="upper left")
plt.show()
#this is a bar chart showing subscribers count for diff types of channel in a particular year
```

```python
df.groupby(["created_year","created_month"])["subscribers"].sum().unstack().head().plot(kind="area",stacked=True)
plt.title("Subscribers growth over time")
plt.xlabel("Year")
plt.ylabel("Subs count")
plt.xticks(rotation=45)
plt.legend(title="Month",bbox_to_anchor=(1.05,1),loc="upper left")
plt.show()
```

```python
#the above graph showed subsriber growth over year's and months too within each year
```

```python
df.groupby(["created_year","Country"])["highest_yearly_earnings"].mean().unstack().head().plot(kind="line",marker="o")
plt.title("trend of highest yearly earnings by start year and country")
plt.xlabel("Created year")
plt.ylabel("Avg highest yearly earnings")
plt.xticks(rotation=45)
plt.legend(title="Country",bbox_to_anchor=(1.05,1),loc="upper left")
plt.show()
```

```python
#the above line graph shows growth of avg highest yearly earnings for each country for 
#each year
```
